## Konu Özeti :









* `print()`: Ekrana veri yazdırmak için kullanılan komuttur.

* `input()`: Kullanıcıdan veri almak için kullanılan komuttur.

* Değişken(Variable): Bir bilgi saklamak için oluşturduğumuz tanımlayıcılara denir.

* Operatör: Tek başlarına herhangi bir anlamı olmayan ancak programın işleyişine katkıda bulunan karakter ya da karakter topluluklarına denir.

* “#” operatörü: Yazılan kodların yanına hatırlatma notları veya yorum yazmak için kullanılan operatördür.

* String – “str” : Karakter dizilerini ifade eden veri tipidir.
Örnek: x = “Python”

* Integer – “int” : Tam sayıları ifade eden veri tipidir.
Örnek: x = 35

* Float – “float” : Ondalıklı sayıları ifade eden veri tipidir.
Örnek: x = 35.5 

* round: Round fonksiyonu parametre ile girilen reel sayıyı ondalıklı kısımdaki değere göre, bir üst veya bir alt tam sayıya belirtilen ondalık sayısı kadar yuvarlamak için kullanılan bir fonksiyondur.

> Bu fonksiyon, x'i ondalık noktasından itibaren n basamak sayısı kadar bir üst veya bir altına döndürür .
-Örnek: 
> print(round(70.73966, 2)) :  70.74
* Toplama: Python’da + operatörü ile toplama işlemi yapabiliriz.
* Çıkarma: Python’da çıkarma işlemi için “-” operatörünü kullanıyoruz.
* Çarpma:Python’da çarpma işlemleri için ” * ” operatörünü kullanıyoruz.
* Python’da bölme işlemi için  ” /  ” bölme işleminden kalanı bulmak için ” % ” kullanmaktayız.

## ÖRNEK 2

Dikdörtgenin iki kenar uzunluğunu kullanıcıya sorup, dikdörtgenin çevresini ve alanını hesaplayan bir program yazınız

Hint : Kenar uzunluklarını input ile alabilirsiniz.

Örnek;

```
>>>input: 
6 , 4

>>>output:
Çevre: 20
Alan: 24 
```





````python
kısa = int(input("Kısa kenar : "))
uzun = int(input("Uzun Kenar : "))
çevre = 2 * (uzun + kısa)
alan = kısa * uzun
print("Dikdörtgenin çevresi :",çevre)
print("Dikdörtgenin alanı : ",alan)
````
